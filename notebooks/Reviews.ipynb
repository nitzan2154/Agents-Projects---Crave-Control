# Shortest reviews

```python
import pandas as pd
import json

df_rev_new["Review_Length"] = df_rev_new["Review"].apply(lambda x: len(x.split()))

# Step 1: Filter Reviews with at Least 7 Words and at Most 21 Words
df_filtered = df_rev_new[(df_rev_new["Review_Length"] >= 7) & (df_rev_new["Review_Length"] <= 30)]

# Step 2: Keep the Two Shortest Reviews for Each Recipe
df_filtered = df_filtered.sort_values(by=["RecipeId", "Review_Length"])  # Sort by RecipeId & review length
df_shortest_two = df_filtered.groupby("RecipeId").head(2)  # Keep only the two shortest per RecipeId

# Step 3: Group the Two Shortest Reviews into One Row Per Recipe
df_grouped = df_shortest_two.groupby("RecipeId")["Review"].apply(lambda x: " ".join(x)).reset_index()
df_grouped["Total_Review_Length"] = df_grouped["Review"].apply(lambda x: sum(len(r.split()) for r in x.split(" ")))  # Sum up word counts

# Step 4: Keep Only the 20,000 Recipes with the Shortest Total Review Length
df_final = df_grouped.nsmallest(20000, "Total_Review_Length")  # Select 20,000 shortest total reviews

# Step 5: Save Recipe IDs of Selected Reviews to JSON
selected_recipe_ids = df_final["RecipeId"].tolist()
with open("selected_recipe_ids.json", "w") as f:
    json.dump(selected_recipe_ids, f)

# Save Final Processed Dataset (Optional)
df_final.to_json("processed_reviews.json", orient="records", indent=4)

print(f" Process Completed! {len(df_final)} recipes selected and saved to 'selected_recipe_ids.json'")
```

```python
async def async_summarize_text(session, text, max_words=12):
    """
    Uses Gemini to summarize a review in up to max_words words.
    Runs asynchronously to allow multiple requests in parallel.

    Args:
        session (aiohttp.ClientSession): The HTTP session.
        text (str): The input review to summarize.
        max_words (int): The maximum number of words in the summary.

    Returns:
        str: The summarized review.
    """
    prompt = f"Summarize the following review in a maximum of {max_words} words while keeping key insights:\n\n{text}\n\nReturn only the summary."

    try:
        # ✅ Add a small delay to avoid API rate limits
        await asyncio.sleep(0.1)

        # Run Gemini API request asynchronously
        response = await asyncio.get_event_loop().run_in_executor(
            None, lambda: model.generate_content(prompt)
        )
        return response.text.strip()
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
async def process_reviews_async(df_grouped_reviews, max_words=12, batch_size=20):
    """
    Processes reviews asynchronously in batches of `batch_size` using Gemini.
    Shows progress using tqdm.

    Args:
        df_grouped_reviews (DataFrame): The filtered & grouped reviews dataset.
        max_words (int): The maximum number of words in each summary.
        batch_size (int): Number of parallel API requests.

    Returns:
        List[str]: List of summarized reviews.
    """
    async with aiohttp.ClientSession() as session:
        results = []
        tasks = []

        # ✅ Use tqdm to show progress
        for i, review in enumerate(tqdm_asyncio(df_grouped_reviews["Review"], desc="🔄 Summarizing Reviews", total=len(df_grouped_reviews))):
            tasks.append(async_summarize_text(session, review, max_words))

            # ✅ When batch size is reached, run tasks and reset
            if len(tasks) >= batch_size:
                results.extend(await asyncio.gather(*tasks))
                tasks = []  # Reset task list

        # ✅ Process any remaining tasks
        if tasks:
            results.extend(await asyncio.gather(*tasks))

        return results
```

```python
async def process_reviews_async(df_grouped_reviews, max_words=12, batch_size=10):
    """
    Processes reviews asynchronously in batches of `batch_size` using Gemini.
    Shows progress using tqdm.

    Args:
        df_grouped_reviews (DataFrame): The filtered & grouped reviews dataset.
        max_words (int): The maximum number of words in each summary.
        batch_size (int): Number of parallel API requests.

    Returns:
        List[str]: List of summarized reviews.
    """
    async with aiohttp.ClientSession() as session:
        results = []
        tasks = []
        total_reviews = len(df_grouped_reviews)

        # ✅ Use tqdm to show progress
        with tqdm(total=total_reviews, desc="🔄 Summarizing Reviews") as pbar:
            for i, review in enumerate(df_grouped_reviews["Review"]):
                tasks.append(async_summarize_text(session, review, max_words))

                # ✅ When batch size is reached, run tasks and reset
                if len(tasks) >= batch_size:
                    print(f"🚀 Processing batch {i // batch_size + 1} of {total_reviews // batch_size}...")  # Debug print
                    results.extend(await asyncio.gather(*tasks))
                    tasks = []  # Reset task list
                    pbar.update(batch_size)  # ✅ Update progress bar

            # ✅ Process any remaining tasks
            if tasks:
                results.extend(await asyncio.gather(*tasks))
                pbar.update(len(tasks))  # ✅ Update progress bar for last batch

        return results
```

```python
# ✅ Load grouped reviews dataset
df_grouped_reviews = pd.read_json("grouped_reviews.json")

# ✅ Run Asynchronous Summarization
loop = asyncio.get_event_loop()
summarized_reviews = loop.run_until_complete(process_reviews_async(df_grouped_reviews, max_words=12, batch_size=20))

# ✅ Store Summarized Reviews in DataFrame
df_grouped_reviews["Summarized_Review"] = summarized_reviews

# ✅ Save Summarized Reviews to JSON
df_grouped_reviews.to_json("summarized_reviews.json", orient="records", indent=4)

print(f"✅ Summarization Completed! Results saved to 'summarized_reviews.json'")
```

```python
import openai
import pandas as pd
import time
# Configure Azure OpenAI API
client = openai.AzureOpenAI(
    api_key=AZURE_OPENAI_API_KEY,
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_version=API_VERSION
)

# Choose Model
AZURE_DEPLOYMENT_NAME = GPT_DEPLOYMENT  # Example: "team1-gpt4o"

def summarize_text(text, max_words=12):
    """
    Simple function to summarize a review using Azure OpenAI API.
    """
    prompt = f"Summarize the following review in a maximum of {max_words} words while keeping key insights:\n\n{text}\n\nReturn only the summary."

    try:
        response = client.chat.completions.create(
            model=AZURE_DEPLOYMENT_NAME,
            messages=[{"role": "user", "content": prompt}],
            max_tokens=50,
            temperature=0.5
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error: {str(e)}")
        # Wait 2 seconds if there's an error (might be rate limiting)
        time.sleep(2)
        return f"Error: {str(e)}"

# Main execution
def main():
    print(f"Starting full dataset summarization at {time.strftime('%H:%M:%S')}")

    # Load Processed Reviews Dataset
    try:
        df_processed_reviews = pd.read_json("processed_reviews.json")
        print(f"Loaded {len(df_processed_reviews)} reviews from dataset")
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return

    # Ensure the Summarized_Review column exists
    if "Summarized_Review" not in df_processed_reviews.columns:
        df_processed_reviews["Summarized_Review"] = ""

    # Check if we have a partial progress file
    try:
        partial_df = pd.read_json("summarized_reviews_partial.json")
        # Count how many reviews already have summaries
        completed_count = partial_df["Summarized_Review"].notna().sum()
        if completed_count > 0:
            print(f"Found existing progress: {completed_count} reviews already summarized")
            # Copy the summaries we already have
            df_processed_reviews.loc[:completed_count-1, "Summarized_Review"] = partial_df.loc[:completed_count-1, "Summarized_Review"]
            start_index = completed_count
        else:
            start_index = 0
    except:
        # No progress file found, start from beginning
        start_index = 0
        print("No existing progress found, starting from the beginning")

    results = list(df_processed_reviews["Summarized_Review"])
    total_reviews = len(df_processed_reviews)
    start_total = time.time()

    # Process each review one at a time, starting from where we left off
    for i in range(start_index, total_reviews):
        review = df_processed_reviews["Review"].iloc[i]
        print(f"\nReview {i+1}/{total_reviews}: Processing...")

        start_time = time.time()
        summary = summarize_text(review, max_words=12)
        elapsed = time.time() - start_time

        results[i] = summary
        print(f"✓ Completed review {i+1} in {elapsed:.2f}s: '{summary}'")

        # Save progress every 50 reviews or at the end
        if (i + 1) % 50 == 0 or i == total_reviews - 1:
            df_processed_reviews["Summarized_Review"] = results
            df_processed_reviews.to_json("summarized_reviews_partial.json", orient="records", indent=4)
            print(f"Saved checkpoint: {i+1}/{total_reviews} reviews")

        # Show time estimate and progress
        time_so_far = time.time() - start_total
        completed_reviews = i + 1 - start_index
        if completed_reviews > 0:  # Avoid division by zero
            avg_time_per_review = time_so_far / completed_reviews
            estimated_total_time = avg_time_per_review * (total_reviews - start_index)
            remaining_time = estimated_total_time - time_so_far

            # Convert remaining time to hours, minutes, seconds for better readability
            hours, remainder = divmod(remaining_time, 3600)
            minutes, seconds = divmod(remainder, 60)

            print(f"Progress: {i+1}/{total_reviews} reviews " +
                  f"({(i+1)/total_reviews*100:.1f}%) | " +
                  f"Avg: {avg_time_per_review:.2f}s per review | " +
                  f"Est. remaining: {int(hours)}h {int(minutes)}m {int(seconds)}s")

    # Save Final Summarized Reviews to JSON
    df_processed_reviews["Summarized_Review"] = results
    df_processed_reviews.to_json("summarized_reviews_full.json", orient="records", indent=4)

    total_time = time.time() - start_total
    # Convert total time to hours, minutes, seconds
    hours, remainder = divmod(total_time, 3600)
    minutes, seconds = divmod(remainder, 60)

    print(f"\nSummarization Completed!")
    print(f"Total time: {int(hours)}h {int(minutes)}m {int(seconds)}s")
    print(f"Results saved to 'summarized_reviews_full.json'")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
import openai
import pandas as pd
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

# Configure Azure OpenAI API
client = openai.AzureOpenAI(
    api_key=AZURE_OPENAI_API_KEY,
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_version=API_VERSION
)

# Choose Model
AZURE_DEPLOYMENT_NAME = GPT_DEPLOYMENT  # Example: "team1-gpt4o"

# Configuration
BATCH_SIZE = 10  # Number of reviews to process in a single API call
MAX_WORKERS = 3  # Number of parallel threads
RESUME_FROM = 451  # Start from review #451

def batch_summarize_reviews(review_batch, max_words=12):
    """
    Batch summarize multiple reviews in a single API call.
    """
    # Create a batch prompt with all reviews
    batch_prompt = "Summarize each of the following reviews in a maximum of " + \
                   f"{max_words} words while keeping key insights. " + \
                   "Format your response as a JSON array of strings, with each string being a summary:\n\n"

    for i, review in enumerate(review_batch):
        batch_prompt += f"Review {i+1}: {review}\n\n"

    batch_prompt += "Return only the JSON array of summaries."

    try:
        response = client.chat.completions.create(
            model=AZURE_DEPLOYMENT_NAME,
            messages=[{"role": "user", "content": batch_prompt}],
            max_tokens=100 * len(review_batch),  # Scale tokens with batch size
            temperature=0.5,
            response_format={"type": "json_object"}  # Request JSON response
        )
        result = response.choices[0].message.content.strip()

        # Parse the JSON response - try different approaches as the model might return different formats
        import json
        try:
            summaries_data = json.loads(result)
            # Handle different possible JSON structures
            if isinstance(summaries_data, list):
                return summaries_data
            elif isinstance(summaries_data, dict) and "summaries" in summaries_data:
                return summaries_data["summaries"]
            elif isinstance(summaries_data, dict):
                # Try to extract an array if it exists in the dictionary
                for key, value in summaries_data.items():
                    if isinstance(value, list):
                        return value
                # Fall back to values if no arrays found
                return list(summaries_data.values())
        except:
            # If JSON parsing fails, return an error message for each review
            return [f"Error parsing JSON response: {result}" for _ in review_batch]

    except Exception as e:
        print(f"Error in batch processing: {str(e)}")
        time.sleep(2)  # Wait if there's an error (might be rate limiting)
        return [f"Error: {str(e)}" for _ in review_batch]

def process_batch_with_retry(batch_reviews, batch_indices, max_retries=3):
    """Process a batch of reviews with retry logic"""
    for attempt in range(max_retries):
        try:
            summaries = batch_summarize_reviews(batch_reviews)
            # Ensure we got the right number of summaries
            if len(summaries) == len(batch_reviews):
                return list(zip(batch_indices, summaries))
            else:
                print(f"Warning: Expected {len(batch_reviews)} summaries but got {len(summaries)}. Retrying...")
                time.sleep(2)
        except Exception as e:
            print(f"Batch processing attempt {attempt+1} failed: {str(e)}")
            time.sleep(3)  # Wait between retries

    # If all retries failed, return error messages
    return [(idx, f"Failed after {max_retries} attempts") for idx in batch_indices]

# Main execution
def main():
    print(f"Starting batch dataset summarization at {time.strftime('%H:%M:%S')}")

    # Load Processed Reviews Dataset
    try:
        df_processed_reviews = pd.read_json("processed_reviews.json")
        print(f"Loaded {len(df_processed_reviews)} reviews from dataset")
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return

    # Ensure the Summarized_Review column exists
    if "Summarized_Review" not in df_processed_reviews.columns:
        df_processed_reviews["Summarized_Review"] = ""

    # Check if we have a partial progress file and load it
    try:
        partial_df = pd.read_json("summarized_reviews_partial.json")
        # Copy the summaries we already have
        for i in range(min(len(partial_df), RESUME_FROM)):
            if pd.notna(partial_df["Summarized_Review"].iloc[i]):
                df_processed_reviews.loc[i, "Summarized_Review"] = partial_df["Summarized_Review"].iloc[i]

        print(f"Loaded existing progress: Will resume from review #{RESUME_FROM}")
    except Exception as e:
        print(f"No valid progress file found: {e}")
        # If no progress file or corrupt file, but user specified RESUME_FROM
        if RESUME_FROM > 0:
            print(f"Warning: No valid progress file but RESUME_FROM = {RESUME_FROM}. Check if this is intended.")

    results = list(df_processed_reviews["Summarized_Review"])
    total_reviews = len(df_processed_reviews)
    start_total = time.time()
    remaining_reviews = total_reviews - RESUME_FROM

    print(f"Processing {remaining_reviews} remaining reviews in batches of {BATCH_SIZE}")

    # Create batches of reviews starting from RESUME_FROM
    batches = []
    batch_indices = []

    for i in range(RESUME_FROM, total_reviews, BATCH_SIZE):
        end_idx = min(i + BATCH_SIZE, total_reviews)
        batch = df_processed_reviews["Review"].iloc[i:end_idx].tolist()
        indices = list(range(i, end_idx))

        batches.append(batch)
        batch_indices.append(indices)

    total_batches = len(batches)
    completed_batches = 0

    # Process batches in parallel
    with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        # Submit all batch jobs
        future_to_batch = {
            executor.submit(process_batch_with_retry, batches[i], batch_indices[i]): i
            for i in range(total_batches)
        }

        # Process results as they complete
        for future in as_completed(future_to_batch):
            batch_num = future_to_batch[future]
            completed_batches += 1

            try:
                batch_results = future.result()
                # Update results with batch summaries
                for idx, summary in batch_results:
                    results[idx] = summary

                # Show batch completion
                print(f"\nBatch {completed_batches}/{total_batches} completed")
                for idx, summary in batch_results:
                    print(f"Review #{idx+1}: '{summary}'")

                # Save progress every 5 batches or at the end
                if completed_batches % 5 == 0 or completed_batches == total_batches:
                    df_processed_reviews["Summarized_Review"] = results
                    df_processed_reviews.to_json("summarized_reviews_partial.json", orient="records", indent=4)
                    print(f"Saved checkpoint: {completed_batches}/{total_batches} batches")

                # Show time estimate and progress
                time_so_far = time.time() - start_total
                if completed_batches > 0:
                    avg_time_per_batch = time_so_far / completed_batches
                    estimated_total_time = avg_time_per_batch * total_batches
                    remaining_time = estimated_total_time - time_so_far

                    # Convert remaining time to hours, minutes, seconds
                    hours, remainder = divmod(remaining_time, 3600)
                    minutes, seconds = divmod(remainder, 60)

                    reviews_completed = sum(len(batch_indices[i]) for i in range(completed_batches))
                    progress_pct = (RESUME_FROM + reviews_completed) / total_reviews * 100

                    print(f"Progress: {RESUME_FROM + reviews_completed}/{total_reviews} reviews " +
                          f"({progress_pct:.1f}%) | " +
                          f"Avg: {avg_time_per_batch:.2f}s per batch | " +
                          f"Est. remaining: {int(hours)}h {int(minutes)}m {int(seconds)}s")

            except Exception as e:
                print(f"Error processing batch {batch_num}: {str(e)}")

    # Save Final Summarized Reviews to JSON
    df_processed_reviews["Summarized_Review"] = results
    df_processed_reviews.to_json("summarized_reviews_full.json", orient="records", indent=4)

    total_time = time.time() - start_total
    # Convert total time to hours, minutes, seconds
    hours, remainder = divmod(total_time, 3600)
    minutes, seconds = divmod(remainder, 60)

    print(f"\nBatch Summarization Completed!")
    print(f"Total time: {int(hours)}h {int(minutes)}m {int(seconds)}s")
    print(f"Results saved to 'summarized_reviews_full.json'")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
import json
import tiktoken
from langchain.embeddings import AzureOpenAIEmbeddings
import time

with open("final_summary.json", "r") as file:
    final_summary = json.load(file)

# Extract relevant columns (RecipeId and Summarized_Review)
df_final_summary = pd.DataFrame(final_summary)
df_embeddings = df_final_summary[['RecipeId', 'Summarized_Review']]

embedding_data = df_embeddings.to_dict(orient='records')

print(embedding_data[:5])
```

```python
# Initialize the tokenizer for the embeddings model
tokenizer = tiktoken.get_encoding("cl100k_base")

# Function to calculate tokens for a single document
def calculate_tokens(text):
    tokens = tokenizer.encode(text)
    return len(tokens)

# Calculate total tokens for all texts in embedding_data
total_input_tokens = sum(calculate_tokens(doc['Summarized_Review']) for doc in embedding_data)
print(f"Total input tokens: {total_input_tokens}")
```

```python
# Function to get embeddings with RecipeId
def get_embeddings_with_review_ids(texts, model, batch_size=10):
    embeddings = []
    failed_batches = []

    for idx in range(0, len(texts), batch_size):
        batch = texts[idx:idx + batch_size]
        batch_recipe_ids = [doc['RecipeId'] for doc in batch]  # Extract RecipeIds for this batch
        batch_texts = [doc['Summarized_Review'] for doc in batch]

        try:
            response = model.embed_documents(batch_texts)

            # Combine RecipeIds and embeddings
            for recipe_id, embedding in zip(batch_recipe_ids, response):
                embeddings.append({'RecipeId': recipe_id, 'Embedding': embedding})

            # Print progress for each batch
            print(f"Batch {idx // batch_size + 1} completed")

        except Exception as e:
            # Store failed batches to retry later if needed
            print(f"Error processing batch {idx // batch_size + 1}: {str(e)}")
            failed_batches.append(batch)

    # If there are failed batches, handle retry
    if failed_batches:
        print(f"\nRetrying {len(failed_batches)} failed batches...")
        time.sleep(5)  # Adding sleep to prevent overloading the API
        return get_embeddings_with_review_ids(failed_batches, model, batch_size)

    return embeddings
```

```python
# Example of processing a small sample of 5 reviews for testing
sample_size = 5  # Small batch size for testing

# Load the final summary data
with open("final_summary.json", "r") as file:
    final_summary = json.load(file)

# Select the first 5 reviews for testing
sample_data = final_summary[:sample_size]

# Extract relevant columns (RecipeId and Summarized_Review)
df_sample = pd.DataFrame(sample_data)[['RecipeId', 'Summarized_Review']]

# Define your embedding function for small batches
def get_embeddings_with_review_ids(texts, model, batch_size=5):
    embeddings = []
    for idx in range(0, len(texts), batch_size):
        batch = texts[idx:idx + batch_size]
        batch_recipe_ids = [doc['RecipeId'] for doc in batch]  # Extract RecipeId for each review
        response = model.embed_documents([doc['Summarized_Review'] for doc in batch])  # Embed the reviews

        # Combine RecipeIds with the embeddings
        for recipe_id, embedding in zip(batch_recipe_ids, response):
            embeddings.append({"RecipeId": recipe_id, "Embedding": embedding})

    return embeddings

# Assuming the model is initialized as emb
embeddings = get_embeddings_with_review_ids(df_sample.to_dict(orient='records'), embedder_gpt_model, batch_size=5)

# Print out the results
for embedding in embeddings:
    print(f"RecipeId: {embedding['RecipeId']} | Embedding: {embedding['Embedding'][:10]}...")  # Show the first few elements of the embedding
```
